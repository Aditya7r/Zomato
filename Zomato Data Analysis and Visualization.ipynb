```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('zomato.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
#dropping columns

df = df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'], axis = 1)
df.head()
```

```python
df.info()
```

```python
#drop duplicates

df.drop_duplicates(inplace=True)
df.shape
```

Cleaning Row column 

```python
#Cleaning Row column 
df['rate'].unique()
```

Removing 'NEW', '-', and '/5' from rate column

```python
#Removing 'NEW', '-', and '/5' from rate column

def clear_rate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/') #splitting '/' in 4.1/5
        value = value[0] #keeping only 4.1 in 4.1/5
        return float(value)

df['rate'] = df['rate'].apply(clear_rate)
df.head()
```

Filling Null Values in rate with Mean

```python
df['rate'].isnull().sum() #number of null values = 10019
```

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

Dropping Null values

```python
df.dropna(inplace=True)
df.info()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

Dropping listed_in(city) because same information is in location column

```python
df = df.drop(['listed_in(city)'],axis=1)
```

```python
df['approx_cost(for two people)'].unique
```

Coverting approx_cost(for two people) into float

```python
def rem_comma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)

df['approx_cost(for two people)'] = df['approx_cost(for two people)'].apply(rem_comma)
df['approx_cost(for two people)'].unique()
```

```python
df['rest_type'].value_counts() #can categorize rest_type with less counts as Others
```

```python
rest_type = df['rest_type'].value_counts(ascending=False)

rest_type_lessthan_1000 = rest_type[rest_type<1000]
rest_type_lessthan_1000
```

Categorize rest_type with less than 1000 counts as Others

```python
def handle_rest_type(value):
    if(value in rest_type_lessthan_1000):
        return 'Others'
    else:
        return value

df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.location.value_counts()
```

Categorize location with less than 300 counts as Others

```python
location = df['location'].value_counts(ascending=False)

location_lessthan_300 = location[location<300]

def handlelocation(value):
    if(value in location_lessthan_300):
        return 'Others'
    else:
        return value


df['location'] = df['location'].apply(handlelocation)
df['location'].value_counts()
```

Categorize Cuisines with less than 100 counts as Others

```python
cuisines = df['cuisines'].value_counts(ascending=False)

cuisines_lessthan_100 = cuisines[cuisines<100]

def handle_cuisines(value):
    if(value in cuisines_lessthan_100):
        return 'Others'
    else:
        return value


df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
